```python
# Drop Genre (gender column)
df_mall.drop(columns=["Genre"], inplace=True)
df_mall.head()
```

```python
# Initializing model with K = 5 (5 clusters)
model = KMeans(n_clusters=5, random_state=5)
model
```

```python
# Fitting model
model.fit(df_mall)
```

```python
# get predictions
predictions = model.predict(df_mall)
print(predictions)
```

```python
# Add a new class column to the df_mall
df_mall["class"] = model.labels_
df_mall.head()
```
